```python
import tensorflow as tf
import tensorflow_datasets as tfds
import cv2
import os
import time
import signal
import multiprocessing
import matplotlib as mpl
from tensorflow import keras
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Activation, Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow.keras.backend as K
from datetime import datetime
from PIL import Image

DATADIR = "datasets/"
LOGDIR = "logs/"
MODELDIR = "models/"

IMG_WIDTH = 96
IMG_HEIGHT = 96
BATCH_SIZE = 32

PCAM, PCAM_INFO = tfds.load("patch_camelyon", with_info=True, as_supervised=True)
figures = tfds.as_dataframe(PCAM['validation'].take(1000), PCAM_INFO)

count = 1
for fig in figures.iterrows():
    data = fig[1].values[0]
    label = fig[1].values[1]
    Image.fromarray(data).save("extracted\\" + str(label) + "\\" + str(count) + ".jpg")
    count = count + 1
```
